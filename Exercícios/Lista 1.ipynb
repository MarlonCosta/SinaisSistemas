### Exercício 2

#### a)

```python
t = np.linspace(-np.pi,0,200)
f = 5
sigma = 0
omega = 2*np.pi*f
s = sigma + 1j*omega
y = np.exp(s*t)
plt.plot(t,y)
```

```python
t = np.linspace(0, 2*np.pi,200)
f = 5
sigma = 0
omega = 2*np.pi*f
s = sigma + 1j*omega
y = np.exp(s*t)
plt.plot(t,y)
```

A função é periódica, então apenas alteramos a janela de observação da função.

#### b)

```python
t = np.linspace(-np.pi,0,200)
f = 5
sigma = 2
omega = 2*np.pi*f
s = sigma + 1j*omega
y = np.exp(s*t)
plt.plot(t,y)
```

```python
t = np.linspace(0, 2*np.pi,200)
f = 5
sigma = 2
omega = 2*np.pi*f
s = sigma + 1j*omega
y = np.exp(s*t)
plt.plot(t,y)
```

Aqui, apesar de a função não ser periódica, também só alteramos a janela da observação da mesma.

A função começa com valor nulo e sua amplitude aumenta a cada oscilação.

Nos dois intervalos observados, temos que de $[-\pi, 0]$ temos valores máximo (1.0) e mínimo (-0.75) bem menores para a amplitude, pois estamos observando um intervalo mais próximo do início da função, portanto, mais próximos de 0. Já de $[0, 2\pi]$ temos valores limite bem maiores, próximos de 200000.

```python
t = np.linspace(-np.pi,0,200)
f = 5
sigma = -2
omega = 2*np.pi*f
s = sigma + 1j*omega
y = np.exp(s*t)
plt.plot(t,y)
```

```python
t = np.linspace(0, 2*np.pi,200)
f = 5
sigma = -2
omega = 2*np.pi*f
s = sigma + 1j*omega
y = np.exp(s*t)
plt.plot(t,y)
```

Neste caso, temos uma função com valor infinito que decai até zero. Novamente, alterando os limites de t, apenas alteramos a janela de observação da mesma.

Neste caso, devida a natureza da função, temos um comportamento oposto ao da função anterior, onde:

Em $[0, \pi]$, temos como valores limite próximos de $[-0.75, 1.00]$

Em $[-\pi, 0]$, os valores limite são próximos de $[-400, 400]$

Já em $[0, 2\pi]$ os valores limite são  próximos de $[-0.75, 1.00]$
