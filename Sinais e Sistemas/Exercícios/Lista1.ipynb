Em $[-\pi, 0]$, os valores limite são próximos de $[-400, 400]$

```python
def porta(t,tau): return 1*(np.abs(t)<=tau/2)
```

```python
t = np.linspace(-1, 1, 500)
y = porta(t, .5)
plt.plot(t,y)
```

### Exercício 12

```python
def square_w_hs(t, tau):
    """Utiliza a função heaviside para produzir a função porta"""
    y0 = np.heaviside(t+tau/2,1)
    y1 = np.heaviside(-t+tau/2,1)
    
    y2 = y0 + y1
    
    return y2
```

```python
t = np.linspace(-1, 1, 500)
y = square_w_hs(t, .5)

plt.plot(t,y)
```

Podemos obter a função porta a partir da soma de dois sinais degrau unitário que sejam simétricos em relação ao eixo y, tendo tau/2 como o ponto de mudança de valor da função.

## Onda quadrada

```python
t = np.linspace(-10, 10, 500)
y = ss.square(t, duty=.5)
plt.plot(t,y)
```

### Exercício 13

### Exercício 14

```python
t = np.linspace(-10, 10, 500)
y = ss.square(t, duty=.5)
y2 = ss.square(t+np.pi/2, duty=.5)

plt.plot(t,y+y2)
```

Aqui, somamos o sinal com ele mesmo deslocado em $\frac{1}{4}$ de período, ou seja, deslocado $\frac{\pi}{2}$.

## Pulso Triangular

```python
def triang(t,tau): return (1 - np.abs(t)/tau)*(np.abs(t)<float(tau))
```

```python
t = np.linspace(-1, 1, 500)
y = triang(t, .5)
plt.plot(t,y)
```

## Sinal Rampa

```python
def rampa(t): return np.maximum(0,t)
```

```python
t = np.linspace(-1, 1, 500)
y = rampa(t)
plt.plot(t,y)
```
